**Melakukan Eksplorasi Data Business Approach**
1. menghitung jumlah order; 
2. menghitung jumlah customer;
3. menghitung jumlah customer yang melakukan order;
4. menghitung jumlah order per order status;
5. menghitung jumlah customer per city

```python
#Menghitung Jumlah order
jumlah_order=tantut_merge_b.order_id.nunique()
print('Jumlah Order:',jumlah_order)
```

```python
#Menghitung Jumlah Customer
jumlah_customer=tantut_merge_b.customer_unique_id.nunique()
jumlah_customer
```

```python
#menghitung jumlah customer yang melakukan order
jumlah_customer_order=tantut_merge_b.customer_id.nunique()
jumlah_customer_order
```

```python
#menghitung jumlah order per order status
order_per_status=tantut_merge_b.groupby('order_status')['order_id'].nunique().reset_index()
order_per_status
```

```python
order_per_status.rename(columns={"order_id": "total_order"}, inplace=True)
order_per_status
```

```python
order_per_status.sort_values(by="total_order", ascending=False)
```

```python
#menghitung jumlah customer per city
customer_per_city=tantut_merge_b.groupby(['customer_city','customer_state'])['customer_unique_id','order_id'].nunique().reset_index()
customer_per_city.rename(columns={"customer_unique_id": "total_customer"}, inplace=True)
customer_per_city.rename(columns={"order_id": "order_total"}, inplace=True)
customer_per_city.sort_values(by="total_customer", ascending=False,inplace=True)
```

```python
customer_city=tantut_merge_b.groupby(['customer_city','payment_type']).agg(
    total_customer=('customer_unique_id','nunique'),
    total_order   =('order_id','nunique'),
    total_payment =('payment_value','sum'),
)
customer_city.sort_values(by="total_payment", ascending=False)
```

Data Transformation
1. Mengubah Tanggal Menjadi tipe dari DD/MM/YYYY HH:MM:SS --> DD/MM/YYYY
2. Membuat CASE WHEN bersyarat

```python
#Mengubah Tanggal Menjadi tipe dari DD/MM/YYYY HH:MM:SS --> DD/MM/YYYY
tantut_merge_b['order_purchase_timestamp'] = pd.to_datetime(tantut_merge_b['order_purchase_timestamp']).dt.date
tantut_merge_b['order_approved_at'] = pd.to_datetime(tantut_merge_b['order_approved_at']).dt.date
tantut_merge_b['order_approved_at'] = pd.to_datetime(tantut_merge_b['order_approved_at']).dt.date
tantut_merge_b['order_delivered_carrier_date'] = pd.to_datetime(tantut_merge_b['order_delivered_carrier_date']).dt.date
tantut_merge_b['order_delivered_customer_date'] = pd.to_datetime(tantut_merge_b['order_delivered_customer_date']).dt.date
tantut_merge_b['order_estimated_delivery_date'] = pd.to_datetime(tantut_merge_b['order_estimated_delivery_date']).dt.date
tantut_merge_b
```

```python
#Case When in Python
tantut_merge_b.loc[(tantut_merge_b['order_status'] == 'delivered'), 'bonus_status'] = 'OK'
tantut_merge_b.loc[(tantut_merge_b['order_status'] == 'unavailable'), 'bonus_status'] = 'NEED TO FU'
tantut_merge_b.loc[(tantut_merge_b['order_status'] == 'canceled'), 'bonus_status'] = 'NEED TO RESEND'
tantut_merge_b.loc[(tantut_merge_b['order_status'] == 'shipped'), 'bonus_status'] = 'Almost OK'
tantut_merge_b.loc[(tantut_merge_b['order_status'] != 'delivered'), 'bonus_status'] = 'NOT OK'
tantut_merge_b
```
